# Opening a New Gym in London

### Introduction
I am creating this report in order to potentially help future companies, both first-time owners and those with existing gyms already, decide what area of London is most suitable for the opening of a new Gym. Whilst there are many gyms currently open within the city of London (800 according to 'https://www.londonlive.co.uk/news/2017-05-21/london-s-booming-fitness-industry'), that still only means that there is 1 gym per 11,000 people given a population size of 9 million. This is far from a completely saturated market, and gym memberships across the UK continue to rise ('https://www.welltodoglobal.com/uk-fitness-industry-worth-5bn-as-number-of-gyms-continues-to-grow/').   

When trying to determine the best location to open a new gym, there are a number of factors to consider: 
- Existing gym locations - areas of high density are likely to already be saturated
- Existing gym locations - areas of low density could be as a result of low demand 
- Cost - floor space rent

    
### Data
In order to assess where would be most suitable, I shall be using the Foursquare API to determine existing gym locations and to evaluate density of gyms for a given Borough. I will also use k-means clustering in order to group areas of London into areas that are similar to each other. This will help determine suitable locations when combined with the gym density data, as if a Borough is found to be similar to another in most aspects, but has a lower gym density, then it is possible that there is an opportunity to open one there. 

In addition, I shall incorporate the cost of floor space into my analysis to see if there are places without existing gyms simply due to economic factors. I shall be using the rent data from the following website ('https://www.oktra.co.uk/insights/how-much-does-london-office-space-cost-in-2020/'). Whilst this contains the price per sq m of rental for office space, I will assume it is reflective of the relative expense of each area. Therefore it will be used more as an indicator of cost, rather than a specific value. The table also contains 2 Grades of office space with a range for each grade. For the purposes of this report, I will conduct analysis using both Grades, but only using the mid-point of each given range. I shall also be making the assumption that the new gym will be 150 sq m in size.
